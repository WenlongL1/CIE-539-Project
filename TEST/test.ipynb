```python
import pandas as pd 
import csv

datas = open('ex7.csv')


import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mu0 = 6      # 
sigma = 1    # 
n = 30
x_bar = 7    # 

#  z
z = (x_bar - mu0) / (sigma / np.sqrt(n))

x = np.linspace(mu0 - 4, mu0 + 4, 1000)
y = stats.norm.pdf(x, mu0, sigma/np.sqrt(n))

plt.plot(x, y)
plt.title("Single Sample Z-Test Illustration")
plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")

# 
plt.fill_between(x, 0, y, where=(x <= mu0 - abs(z)*sigma/np.sqrt(n)) | (x >= mu0 + abs(z)*sigma/np.sqrt(n)), color='red', alpha=0.3, label='two-sided p')
# 
plt.fill_between(x, 0, y, where=(x >= x_bar), color='blue', alpha=0.3, label='right-tail p')
# 
plt.fill_between(x, 0, y, where=(x <= x_bar), color='green', alpha=0.3, label='left-tail p')

plt.axvline(mu0, color='black', linestyle='--', label='H0 mean')
plt.axvline(x_bar, color='orange', linestyle='--', label='Sample mean')
plt.legend()
plt.show()
```

```python
import gurobipy as gp
from gurobipy import GRB

# 
m = gp.Model("simple_lp")

# x, y lb=0
vars = m.addVars(['x', 'y'], lb=0, vtype=GRB.CONTINUOUS)

# 
m.addConstr(5*vars['x'] + 2*vars['y'] <= 40, "c0")

# 
m.setObjective(10*vars['x'] + 5*vars['y'], GRB.MAXIMIZE)

# 
m.optimize()

# 
for v in vars.values():
    print(f"{v.VarName} = {v.X}")
print(f"Objective value = {m.ObjVal}")
```
